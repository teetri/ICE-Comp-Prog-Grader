```python
class piggybank:
    def __init__(self):
      self.c1 = 0
      self.c2 = 0
      self.c5 = 0
      self.c10 = 0
        
    def add1(self, n):
        self.c1 += n
    
    def add2(self, n):
        self.c2 += n
    
    def add5(self, n):
         self.c5 += n
    
    def add10(self, n):
        self.c10 += n
    
    def __int__(self):
        return self.c1 + self.c2*2 + self.c5*5 + self.c10*10
    
    def __lt__(self, rhs):
        return int(self) < int(rhs) 
    
    def __str__(self):
        return "{" + "1:" + str(self.c1) + ", " + "2:" + str(self.c2) + ", " + "5:" + str(self.c5) + ", " + "10:" + str(self.c10) + "}" 


p1 = piggybank()
print(int(p1))
p1.add1(10)
print(int(p1))
p1.add2(5)
print(int(p1))
p1.add5(2)
print(int(p1))
p1.add10(1)
print(int(p1))

p2 = piggybank()
p2.add10(5)
print(int(p2))

p1 < p2
print(p1)
print(p2)
```
