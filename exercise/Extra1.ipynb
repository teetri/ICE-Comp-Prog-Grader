```python
n = int(input())
de = dict()
for i in range(n):
  d, seats = input().split()
  de[d] = int(seats)

rank = []
info = dict()
num_s = int(input())
for i in range(num_s):
  x = input().split()
  id = int(x[0])
  score = float(x[1])
  chosen_de = x[2::]
  rank.append([score, id])
  info[id] = chosen_de
rank.sort(reverse= True)

result = dict()
for s in rank:
  for d in info[s[1]]:
    if de[d] > 0 :
      result[s[1]] = d
      de[d] -= 1
      break
sorted_result = sorted([[s, result[s]]for s in result])
for s in sorted_result:
  print(str(s[0]), s[1])
    
'''
5
CP 1
ME 2
PE 1
CHE 1
MT 3
6

59301234521 23.6 PE CP MT CHE
59300799921 44.5 ME CP CHE PE
59300081621 37 PE CHE MT CP
59300653521 61.2 PE MT CP ME
59300002121 19.4 CHE CP ME CP
59300048721 7 ME CP CHE MT
'''
```
