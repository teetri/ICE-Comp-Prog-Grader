```python
import numpy as np 
 
def scale(img, c) :
  x = np.zeros((img.shape[0], int(img.shape[1]/c)))
  for i in range(c):
    x += img[::,i::c]
  y = np.zeros((int(x.shape[0]/c), x.shape[1]))
  for i in range(c):
    y += x[i::c]
  return y/(c**2)

def read_img(): 
  row, col = [int(e) for e in input().split()] 
  img = np.ndarray((row,col)) 
  for i in range(row): 
    img[i] = [float(e) for e in input().split()] 
  return img 
 
def show_output(out): 
  for i in range(out.shape[0]): 
    print(" ".join([str(e) for e in out[i]])) 
```

```python
img = read_img() 
c = int(input()) 
out = scale(img, c) 
show_output(out) 
```
